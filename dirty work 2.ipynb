```python
import pandas as pd
import numpy as np
```

```python
player_info = 'People.csv'
batting = 'Batting.csv'
pitching = 'Pitching.csv'
```

```python
df = pd.read_csv(player_info)
bat_df = pd.read_csv(batting)
pitching_df = pd.read_csv(pitching)
temp_df=pd.DataFrame.merge(df,bat_df,on='playerID')
merged_df=pd.DataFrame.merge(temp_df,pitching_df,on='playerID')

df_2000=merged_df
# [(merged_df['yearID_x']>1999) | (merged_df['yearID_y']>1999)]

len(merged_df)
```

```python
df1 = df_2000[['playerID','birthCity','birthCountry','birthState','nameFirst','nameLast','yearID_x','yearID_y']]
df2 = df1.drop_duplicates(subset='playerID', keep='first')
df2 = df2.replace(np.nan, '', regex=True)
df2.head()
# PR_df = df2[df2['birthCountry']=='P.R.']

```

```python
birth_city_df = df2.groupby(['birthCity','birthCountry','birthState'],as_index=False).count()
birth_city_df.columns=('Birth City','Birth Country', 'Birth State','Count','First Name','Last Name', 'yearID_x','yearID_y')
birth_city_df = birth_city_df.sort_values(by='Count',ascending=False).reset_index(drop=True)

pd.set_option("display.max_rows", None, "display.max_columns", None)

foreign_df = birth_city_df[birth_city_df['Birth Country']!='USA']
usa_df = birth_city_df[birth_city_df['Birth Country']=='USA']
usa_df=usa_df.reset_index(drop=True)


birth_state = usa_df['Birth State'].value_counts()
foreign_players = foreign_df.groupby('Birth Country')['Count'].sum()
foreign_players_df=pd.Series.to_frame(foreign_players)
foreign_players_df.reset_index(level=0, inplace=True)
foreign_players_df.sort_values(by='Count',ascending=False).reset_index(drop=True)

usa_players = usa_df.groupby('Birth State')['Count'].sum()
usa_df=pd.Series.to_frame(usa_players)
usa_df.reset_index(level=0, inplace=True)
usa_df = usa_df.sort_values(by='Count',ascending=False).reset_index(drop=True)
```
